```python
import cv2

image = cv2.imread("image.png")
```

```python
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow("origin", image)
cv2.imshow("hsv",img_hsv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

分离RGB三通道

```python
b, g, r = cv2.split(image)
image_merged = cv2.merge([b,g,r])
```

```python
cv2.imshow("Red", r)
cv2.imshow("Green", g)
cv2.imshow("Blue", b)
cv2.imshow("Merged", image_merged)
cv2.imshow("origin", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

分离HSV通道

```python
H, S, V = cv2.split(img_hsv)
cv2.imshow('H', H)
cv2.imshow('S', S)
cv2.imshow('V', V)
cv2.imshow("hsv",img_hsv)

cv2.imshow("origin", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

rgb三个通道画三维图

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python

fig = plt.figure(figsize=(10, 5))

# 绘制R通道的三维图
ax_r = fig.add_subplot(131, projection='3d')
X = np.arange(0, image.shape[1], 1) #生成x轴数据列表 图片宽度
Y = np.arange(0, image.shape[0], 1)
X, Y = np.meshgrid(X, Y) #对x，y数据执行网格化
ax_r.plot_surface(X, Y, r, cmap='Reds')

# 绘制G通道的三维图
ax_g = fig.add_subplot(132, projection='3d')
ax_g.plot_surface(X, Y, g, cmap='Greens')

# 绘制B通道的三维图
ax_b = fig.add_subplot(133, projection='3d')
ax_b.plot_surface(X, Y, b, cmap='Blues')

plt.show()
```
