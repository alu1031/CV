```python
import cv2
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
```

```python
image = cv2.imread('image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 添加中文文字
font_path = 'simhei.ttf' 
font_size = 80
font = ImageFont.truetype(font_path, font_size)
```

```python
text = '21122897-王子棠'
text_color = (255, 255, 255)
text_position = (50, 50)
text_alignment = 'left'
```

```python
image_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # 将OpenCV图像转换为Pillow图像
draw = ImageDraw.Draw(image_pil)  # 创建绘制对象
draw.text(text_position, text, font=font, fill=text_color, align=text_alignment)  # 绘制文本
```

```python
plt.imshow(image_pil)
plt.show()
```

```python
output_path = 'output.jpg'  # 输出图像路径
image_pil.save(output_path)  # 保存图像
```
